```python

# Creating a 3x2 grid of subplots

fig, axs = plt.subplots(3, 2, figsize=(16, 12), sharex=True)

# Plotting time series

sns.lineplot(data=df, x='Date', y='GDP Growth (%)', color='orange', ax=axs[0, 0]).set_title("GDP Growth Over Time")
sns.lineplot(data=df, x='Date', y='Inflation Rate (%)', color='orange', ax=axs[0, 1]).set_title("Inflation Rate Over Time")
sns.lineplot(data=df, x='Date', y='Unemployment Rate (%)', color='orange', ax=axs[1, 0]).set_title("Unemployment Rate Over Time")
sns.lineplot(data=df, x='Date', y='Crude Oil Price (USD per Barrel)', color='orange', ax=axs[1, 1]).set_title("Crude Oil Prices Over Time")
sns.lineplot(data=df, x='Date', y='Gold Price (USD per Ounce)', color='orange', ax=axs[2, 0]).set_title("Gold Prices Over Time")

# Removing the unused last subplot

axs[2, 1].remove() 

# Centering the last plot (Gold Price) by changing its position to span both bottom columns

from matplotlib.gridspec import GridSpec
gs = axs[2, 0].get_subplotspec().get_gridspec()
axs[2, 0].remove()  # Remove current placement

# Adding a new axis that spans both columns in the last row

ax_centered = fig.add_subplot(gs[2, :])
sns.lineplot(data=df, x='Date', y='Gold Price (USD per Ounce)', color='orange', ax=ax_centered).set_title("Gold Prices Over Time (Centered)")

# Layout & title

plt.suptitle("Key Economic Indicators Over Time", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python

# Features and targets

features = ['Inflation Rate (%)', 'Unemployment Rate (%)', 'Interest Rate (%)',
            'Consumer Confidence Index', 'Retail Sales (Billion USD)', 'Crude Oil Price (USD per Barrel)']

# GDP Growth Prediction

X_gdp = df[features]
y_gdp = df['GDP Growth (%)']
X_train_gdp, X_test_gdp, y_train_gdp, y_test_gdp = train_test_split(X_gdp, y_gdp, test_size=0.2, random_state=42)
model_gdp = LinearRegression().fit(X_train_gdp, y_train_gdp)
y_pred_gdp = model_gdp.predict(X_test_gdp)

# Inflation Prediction

y_inflation = df['Inflation Rate (%)']
X_train_inf, X_test_inf, y_train_inf, y_test_inf = train_test_split(X_gdp, y_inflation, test_size=0.2, random_state=42)
model_inf = LinearRegression().fit(X_train_inf, y_train_inf)
y_pred_inf = model_inf.predict(X_test_inf)

# Evaluation

gdp_r2 = r2_score(y_test_gdp, y_pred_gdp)
gdp_rmse = mean_squared_error(y_test_gdp, y_pred_gdp, squared=False)

inf_r2 = r2_score(y_test_inf, y_pred_inf)
inf_rmse = mean_squared_error(y_test_inf, y_pred_inf, squared=False)

# Coefficients for feature importance

feature_importance = pd.DataFrame({
    'Feature': features,
    'GDP Coefficient': model_gdp.coef_,
    'Inflation Coefficient': model_inf.coef_
})

(gdp_r2, gdp_rmse, inf_r2, inf_rmse, feature_importance)
```

```python
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()

# Use quarterly GDP growth for time series
df_ts = df[['Date', 'GDP Growth (%)']].copy()
df_ts.set_index('Date', inplace=True)
df_ts = df_ts.asfreq('Q')  # Ensure quarterly frequency
df_ts = df_ts.fillna(method='ffill')  # Fill any missing values

# Fit ARIMA model (automated ARIMA is not available, trying manual (1,1,1))
model_arima = ARIMA(df_ts, order=(1, 1, 1))
model_fit = model_arima.fit()

# Forecast next 8 quarters (2 years)
forecast_steps = 8
forecast = model_fit.forecast(steps=forecast_steps)

# Create date index for forecast
forecast_index = pd.date_range(start=df_ts.index[-1] + pd.offsets.QuarterEnd(1), periods=forecast_steps, freq='Q')
forecast_series = pd.Series(forecast, index=forecast_index)

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df_ts, label='Historical GDP Growth')
plt.plot(forecast_series, label='Forecasted GDP Growth', linestyle='--', color='red')
plt.title('üìà GDP Growth Forecast (ARIMA)')
plt.xlabel('Date')
plt.ylabel('GDP Growth (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

forecast_series
```

```python
# Function to forecast and plot ARIMA for a given column
def forecast_arima(df, col_name, steps=8):
    df_series = df[['Date', col_name]].copy()
    df_series.set_index('Date', inplace=True)
    df_series = df_series.asfreq('Q')
    df_series = df_series.fillna(method='ffill')

    model = ARIMA(df_series, order=(1, 1, 1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=steps)
    forecast_index = pd.date_range(start=df_series.index[-1] + pd.offsets.QuarterEnd(1), periods=steps, freq='Q')
    forecast_series = pd.Series(forecast, index=forecast_index)

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(df_series, label=f'Historical {col_name}')
    plt.plot(forecast_series, label=f'Forecasted {col_name}', linestyle='--')
    plt.title(f'üìâ {col_name} Forecast (ARIMA)')
    plt.xlabel('Date')
    plt.ylabel(col_name)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    return forecast_series

# Forecast for Unemployment Rate, Retail Sales, Crude Oil Price
forecast_unemployment = forecast_arima(df, 'Unemployment Rate (%)')
forecast_retail_sales = forecast_arima(df, 'Retail Sales (Billion USD)')
forecast_oil_price = forecast_arima(df, 'Crude Oil Price (USD per Barrel)')

forecast_unemployment, forecast_retail_sales, forecast_oil_price
```
